```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/nva.filteredferq_nostopwords.csv', nrows=20000)
df.head()
```

```python
# trying on prepocessed data , removing stop words and all-MiniLM-L6-v2'
# Import necessary libraries

from bertopic import BERTopic
from sentence_transformers import SentenceTransformer



# Filter the dataset where labels column equals 0
filtered_df = df[df['label'] == 0]

texts = filtered_df['filtered_text_no_stopwords'].dropna().tolist()

# Load a pretrained sentence transformer model for generating document embeddings
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for the dataset
embeddings = embedding_model.encode(texts, show_progress_bar=True)

# Initialize BERTopic with the pretrained embeddings
topic_model = BERTopic(embedding_model=embedding_model, calculate_probabilities=True)

# Fit the BERTopic model on the embeddings
topics, probabilities = topic_model.fit_transform(texts)

# Display topics
print(topic_model.get_topic_info())
```

```python

# trying to reduce number of topics
# Import necessary libraries
from bertopic import BERTopic
from sentence_transformers import SentenceTransformer
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

# Filter the dataset where labels column equals 0
filtered_df = df[df['label'] == 0]

texts = filtered_df['filtered_text_no_stopwords'].dropna().tolist()

# Load a pretrained sentence transformer model for generating document embeddings
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for the dataset
embeddings = embedding_model.encode(texts, show_progress_bar=True)

# Initialize BERTopic with the pretrained embeddings
topic_model = BERTopic(embedding_model=embedding_model, calculate_probabilities=True)

# Fit the BERTopic model on the embeddings
topics, probabilities = topic_model.fit_transform(texts)

# Display initial topics
print("Initial Topics:")
print(topic_model.get_topic_info())

# Reduce the number of topics
desired_topic_count = 50  # Set the desired number of topics (e.g., 50)
topic_model.reduce_topics(texts, nr_topics=desired_topic_count)

# Display reduced topics
print("\nReduced Topics:")
print(topic_model.get_topic_info())

# Calculate coherence score after topic reduction
def extract_topic_words(topic_model, top_n=10):
    topics = topic_model.get_topics()
    topic_words = [
        [word for word, _ in topics[topic_id][:top_n]]
        for topic_id in topics.keys() if topic_id != -1  # Exclude outlier topics
    ]
    return topic_words

def calculate_coherence_score(topic_model, texts, top_n=10):
    tokenized_texts = [text.split() for text in texts]
    dictionary = Dictionary(tokenized_texts)
    topic_words = extract_topic_words(topic_model, top_n=top_n)
    coherence_model = CoherenceModel(
        topics=topic_words,
        texts=tokenized_texts,
        dictionary=dictionary,
        coherence='c_v'
    )
    return coherence_model.get_coherence()

# Calculate coherence score for reduced topics
reduced_coherence_score = calculate_coherence_score(topic_model, texts)
print(f"\nCoherence Score after Reduction: {reduced_coherence_score}")
```

```python
#trying top2vec method :
# Import necessary libraries
from top2vec import Top2Vec
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

# Step 1: Filter the dataset where labels column equals 0
filtered_df = df[df['label'] == 0]

# Extract text data
texts = filtered_df['filtered_text_no_stopwords'].dropna().tolist()

# Step 2: Initialize and fit the Top2Vec model
print("Fitting Top2Vec model...")
top2vec_model = Top2Vec(
    documents=texts,
    embedding_model='all-MiniLM-L6-v2',  # Use a transformer-based embedding model
    speed="deep-learn",                 # Options: "fast-learn", "learn", "deep-learn"
    verbose=True                        # Enables detailed logging
)

# Display the number of topics found
print(f"\nNumber of topics found: {top2vec_model.get_num_topics()}")

# Step 3: Extract topics and their top words
def display_topics(top2vec_model, top_n=10):
    # Get topics from the model
    topic_words, word_scores, topic_scores = top2vec_model.get_topics()  # Adjusted to match the latest Top2Vec
    print("\nTopics:")
    for i, words in enumerate(topic_words):
        print(f"Topic {i}: {', '.join(words[:top_n])}")

# Display the topics
display_topics(top2vec_model)

# Step 4: Calculate coherence score
def extract_top2vec_topic_words(top2vec_model, top_n=10):
    topic_words, _, _ = top2vec_model.get_topics()  # Only retrieve topic words
    return [words[:top_n] for words in topic_words]

def calculate_coherence_score_top2vec(top2vec_model, texts, top_n=10):
    tokenized_texts = [text.split() for text in texts]  # Tokenize texts
    dictionary = Dictionary(tokenized_texts)           # Create a Gensim dictionary
    topic_words = extract_top2vec_topic_words(top2vec_model, top_n=top_n)
    coherence_model = CoherenceModel(
        topics=topic_words,
        texts=tokenized_texts,
        dictionary=dictionary,
        coherence='c_v'
    )
    return coherence_model.get_coherence()

# Calculate and display coherence score
coherence_score = calculate_coherence_score_top2vec(top2vec_model, texts)
print(f"\nCoherence Score: {coherence_score}")

```
